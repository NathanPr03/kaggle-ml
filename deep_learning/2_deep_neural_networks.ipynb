# Deep Networks
So a single layer by itself obviously isn't going to do much good. What we need is to stack multiple layers on top of each other.

## Layers
Layers don't necessarily need to be like the one we displayed earlier. A layer in a NN can be things like feature engineering, simple arithmetic, or a 'proper' kind of layer like a CNN or RNN

## Activation
However, if we just stack linear layers on top of each other... what we will come out with will bea linear model. Data is rarely so simpel that a linear model can accurately 'grasp' its complexities. This where we need activation functions, which 'bend' the model to be non-linear*

A RElu simply sets the floor for a value to be 0. So if a neuron outputs -0.25, it gets 'floored' to 0. Hte RElu's are added onto the end of the neuron, before the next layer

*How does it actually bend? In reality, it just creates a kink so instead of a straight line being plotted we get a line that's straight up until output >= 0 and then is'curved' (still a right angle) while the output >=0



```python
import keras
import keras.layers as layers

# Create dense model with RElu
model = keras.Sequential([
    # the hidden ReLU layers
    layers.Dense(units=512, activation='relu', input_shape=[8]),
    layers.Dense(units=512, activation='relu'),
    layers.Dense(units=512, activation='relu'),
    # the linear output layer
    layers.Dense(units=1),
])

# Create model with separate activation layer (functionally the same thing)
model = keras.Sequential([
    layers.Dense(32, input_shape=[8]),
    layers.Activation('relu'),
    layers.Dense(32),
    layers.Activation('relu'),
    layers.Dense(1),
])
```
